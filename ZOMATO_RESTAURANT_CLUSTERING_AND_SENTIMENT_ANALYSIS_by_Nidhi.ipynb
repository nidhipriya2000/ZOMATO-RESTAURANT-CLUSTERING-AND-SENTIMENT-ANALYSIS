```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# extracting the stopwords from nltk library
sw = stopwords.words('english')
```

```python
def stopwords(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
# Removing stopwords from Cuisines
meta_df['Cuisines'] = meta_df['Cuisines'].apply(lambda text: stopwords(text))
meta_df['Cuisines'].head()
```

```python
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    # replacing the punctuations with no space, 
    # which in effect deletes the punctuation marks 
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
```

```python
#removing punctuation
meta_df['Cuisines'] = meta_df['Cuisines'].apply(lambda x: remove_punctuation(x))
meta_df['Cuisines'].head()
```

```python
#importing regular expression
import re
```

```python
#function for removing repeating character
def cleaning_repeating_char(text):
    return re.sub(r'(.)1+', r'1', text)
```

```python
#applying func cleaning_repeating_char
meta_df['Cuisines'] = meta_df['Cuisines'].apply(lambda x: cleaning_repeating_char(x))
meta_df['Cuisines'].head()
```

```python
#function for cleaning numbers
def cleaning_numbers(data):
    return re.sub('[0-9]+', '', data)
```

```python
#Lemmatization is the grouping together of different forms of the same word
def lemmatization_(text):
  for index,x in enumerate(text):
    doc = nlp(x)  
    l=list()
    for word in doc:
        l.append(word.lemma_)
    text[index]=' '.join(l)
  return text
```

#SENTIMENTAL ANALYSIS

```python
#TextBlob is a Python (2 and 3) library for processing textual data
from textblob import TextBlob
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import plotly.express as px
```

```python
#Creating a function to get subjectivity
def subjectivity(text): 
    return TextBlob(text).sentiment.subjectivity
```

```python
#Creating a function to get polarity
def polarity(text): 
    return TextBlob(text).sentiment.polarity
```

```python
#Creating two separate columns for polarity and subjectivity
review_df['Subjectivity'] = review_df['Review'].apply(subjectivity)
review_df['Polarity'] = review_df['Review'].apply(polarity)
```

```python
#Creating a function for computing negative, neutral and positive analysis
def getAnalysis(score):
    if score <0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
#applying getAnalysis
review_df['Analysis'] = review_df['Polarity'].apply(getAnalysis)
```

```python
#ploting the polarity and subjectivity
fig = px.scatter(review_df, 
                 x='Polarity', 
                 y='Subjectivity', 
                 color = 'Analysis',
                 size='Subjectivity')
```

```python
#adding a vertical line at x=0 for Netural Reviews
fig.update_layout(title='Sentiment Analysis',
                  shapes=[dict(type= 'line',
                               yref= 'paper', y0= 0, y1= 1, 
                               xref= 'x', x0= 0, x1= 0)])
fig.show()
```

```python
# Converting Analysis into numeric form i.e Postive=0 and Negative =1
def sentiment(analysis):
  if analysis == 'Positive':
    return 0
  elif analysis == 'Neutral' :
    return 0
    # positive sentiment
  else:
    return 1
    # neagative sentiment
```

```python
#applying sentiment
review_df['Analysis']=review_df['Analysis'].apply(lambda x:sentiment(x))
```

```python
def model_details(y_train, y_test, y_pred_train, y_pred_test):
  # Accuracy
  print('-'*50)
  train_accuracy = accuracy_score(y_train,y_pred_train)
  print('Train accuracy is:',train_accuracy)
  test_accuracy = accuracy_score(y_test,y_pred_test)
  print('Test accuracy is:',test_accuracy)
  print('-'*50)
  confusion_mat = confusion_matrix(y_pred_test, y_test)
  print("The confusion matrix on the test data is :")
  print(confusion_mat)
  print('-'*50)
  classification_rep= classification_report(y_pred_test, y_test)
  
  print("The classification report on the train data is :")
  print(classification_rep)
```
